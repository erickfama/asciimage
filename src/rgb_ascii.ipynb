# RGB to ASCII image

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read image 
img = Image.open("../image/matrix.jpg")
```

```python
# Display image
plt.imshow(img)
```

```python
# Convert image to grayscale
img_grayscale = img.convert("LA")

# Numpy array
img_grayscale_arr = np.array(img_grayscale)

# Show 
plt.imshow(img_grayscale)
```

```python
neo = img_grayscale.crop((370, 100, 500, 250))
neo
```

```python
print("Full image shape:", img_grayscale_arr.shape)
print("Crop image shape:", neo.size)
```

```python
# Get X and Y max coordinates
max_y = neo.height
max_x = neo.width

print("Max Y:", max_y, "Max X:", max_x)
```

```python
# Define ascii to replace channel colors
light = ["'", ".", ",", "-", "_", '"']
med = ["^", "~", ":", ";", "*", "c"]
dark = ["+", "L", "2", "3", "6", "@"]

lightness = light + med + dark

unit = 255/18
segments = [unit * n for n in range(1, 19)]

# Set gray value to ascii
def is_light(i):
    count = 0
    for seg in segments:
        if i > seg:
            count += 1
    return lightness[count]

def img_to_ascii(image):
    img_gray = image.convert("LA")
    
    max_x = img_gray.width
    max_y = img_gray.height
    
    image_str = ""
    for y in range(0, max_y):
        for x in range(0, max_x):
            i = img_gray.getpixel((x, y))[0]
            ascii_value = is_light(i)
            image_str = image_str + ascii_value
        image_str = image_str + "\n"
    print(image_str)
```

```python
img_to_ascii(img)
```

```python
img_to_ascii(img.crop((210, 900, 630, 960)))
```
